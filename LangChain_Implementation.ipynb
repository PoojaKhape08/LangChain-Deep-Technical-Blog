# LangChain Deep Technical Implementation

This notebook demonstrates core components of LangChain including PromptTemplate, Chains, Memory, and Agents.

```python
# Install and Imports
!pip install langchain langchain-openai

from langchain_core.prompts import PromptTemplate
```

```python
# Prompt Template Example

template = "Explain {topic} in simple terms"

prompt = PromptTemplate(
    input_variables=["topic"],
    template=template
)

output = prompt.format(topic="Machine Learning")
print(output)
```

```python
# Simulated LLM (No API required)

def fake_llm(prompt):
    return f"[LLM Output]: {prompt} is an important concept in AI."

response = fake_llm("Artificial Intelligence")
print(response)
```

```python
# Simulated Chain

def simple_chain(topic):
    formatted_prompt = prompt.format(topic=topic)
    return f"[Chain Output]: {formatted_prompt}"

print(simple_chain("Deep Learning"))
```

```python
# Simulated Memory

memory = []

def save_memory(user_input, bot_output):
    memory.append({"input": user_input, "output": bot_output})

def load_memory():
    return memory

# Storing conversation
save_memory("Hello", "Hi")
save_memory("How are you?", "I am fine")

# Retrieving memory
print(load_memory())
```

```python
# Simulated Agent

def fake_agent(query):
    return f"[Agent]: Deciding best action for '{query}'"

print(fake_agent("Explain Python"))
```

```python
# Combined Flow Demo

topic = "Neural Networks"

formatted = prompt.format(topic=topic)
llm_output = fake_llm(topic)
chain_output = simple_chain(topic)

print("Prompt:", formatted)
print("LLM:", llm_output)
print("Chain:", chain_output)
```

## ✅ Conclusion

This notebook demonstrates how LangChain concepts can be implemented in a modular way.

- PromptTemplates help structure input
- Chains help organize workflow
- Memory maintains context
- Agents enable decision making
